```python
import pandas as pd
from pathlib import Path

DATA_DIR = Path("datasets")
EVENTS_DIR = DATA_DIR.joinpath("events")
TRANSFERS_DIR = DATA_DIR.joinpath("transfers")


events_df = pd.read_csv(EVENTS_DIR.joinpath("events.csv"))
game_info_df = pd.read_csv(EVENTS_DIR.joinpath("ginf.csv"))
event_game_info_df = pd.merge(events_df, game_info_df, on='id_odsp', how='inner')
transfers_df = pd.read_csv(TRANSFERS_DIR.joinpath("transfers.csv"))
```

```python
player_name = 'Zlatan Ibrahimovic'
```

```python
player_event_game_info = event_game_info_df[event_game_info_df['player'] == player_name.lower()]
player_event_game_info
```

```python
player_transfers = transfers_df[transfers_df['player_name'] == player_name]
player_transfers
```

```python
player_path_name = player_name.lower().replace(' ', '_')
player_event_game_info.to_json(EVENTS_DIR.joinpath(player_path_name).with_suffix('.json'))
player_transfers.to_json(TRANSFERS_DIR.joinpath(player_path_name).with_suffix('.json'))
```
